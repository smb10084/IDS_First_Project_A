```python
import numpy as np
from scipy import stats
alpha =  0.005
alpha_var = 0.005

# 1) Are movies that are more popular (operationalized as having more ratings) rated higher than movies that are less popular? [Hint: You can do
# a median-split of popularity to determine high vs. low popularity movies]
"""
Do:
I calculated the popularity of each movie as the number of non missing ratings
and its mean rating. Using a median split on popularity, I divided the 400 movies
into High popularity (n=200) and Low popularity (n=200) samples. Levene's test
was used to check for homogeneity of variances. Followed by an independent samples
Welch's t-test since the variances were hetrogeneous.
Why:
The goal was to determine whether more popular movies(those with more ratings)
are rated significantly higher on average than less popular movies. An indpendent
t-test (Welch's test) was used because it compares the mean of two independent
samples on a continuous variable(average rating). One tail test was used because
we are testing if more popular movies are rated higher not whether there is any
difference in the ratings.
Find:
Lavene's test indicated hetrogeneity of variances p = 0.0004 < 0.005 so Welch's
t-test was applied.
Results showed a large positive t = 17.756 with df=378.54 and with a one-tail p = 4.77x10^-52 <  = 0.005
mean ratings were 2.87 (high popularity) vs. 2.40 (low popularity).
Answer:
at  = 0.005 the result is significant, so our H is droped. this means that movies
with higher popularity are rated significantly higher on average than less ones.
"""

median_popularity = movie_stats["Popularity"].median()
movie_stats["PopularityGroup"] = np.where(movie_stats["Popularity"] >= median_popularity, "High", "Low")
movie_stats["PopularityGroup"].value_counts()

#create two samples
high_means = movie_stats.loc[movie_stats["PopularityGroup"]=="High", "Mean Rating"]
low_means  = movie_stats.loc[movie_stats["PopularityGroup"]=="Low",  "Mean Rating"]

s1, s2 = high_means.size, low_means.size
m1, m2 = high_means.mean(), low_means.mean()
sd1, sd2 = high_means.std(ddof=1), low_means.std(ddof=1)

print("High n =", s1, " | mean =", m1, " | sd =", sd1)
print("Low  n =", s2,  " | mean =", m2,  " | sd =", sd2)


# Levenes test for equality of variances
W, p_levene = stats.levene(high_means, low_means, center='median')
if p_levene >= alpha_var:
    print(f"p = {p_levene:.4f}  {alpha_var}  Homogeneity of variances assumed (use Students t-test)")
    equal_var_flag = True
else:
    print(f"p = {p_levene:.4f} < {alpha_var}  Heterogeneity of variances (use Welchs t-test)")
    equal_var_flag = False

#H : high_means  low_means
#H : high_means > low_means
# Welchs t-test
""" * does not assume homogeneity of variance
    * automatically calculates different degrees of freedom (df) for the two groups
    """
# Degrees of freedom
df_welch = ((sd1**2/s1 + sd2**2/s2)**2) / (((sd1**2/s1)**2)/(s1-1) + ((sd2**2/s2)**2)/(s2-1))
print(f"Welchs df = {df_welch:.2f}")
# If there are missing values (NaN), ignore them and continue the calculation without stopping "nan_policy='omit'"
t_stat, p_val = stats.ttest_ind(high_means, low_means, equal_var=equal_var_flag, nan_policy='omit')
# convert P-value to one-tailed (High > Low)
p_val_one = p_val/ 2 if t_stat > 0 else 1.0
decision = "Significant(drop H0)" if p_val_one < alpha else "Not significant (keep assumption that H0 is true)"
print(f"t = {t_stat:.3f}, p(one-tailed) = {p_val_one:.4g}, alpha = {alpha}")
print("Decision:", decision)



import matplotlib.pyplot as plt

data = [high_means, low_means]
labels = ['High Popularity', 'Low Popularity']

plt.figure(figsize=(7,5))
plt.boxplot(data, tick_labels=labels, patch_artist=True)
plt.title("Comparison of Mean Ratings by Popularity Group", fontsize=14)
plt.ylabel("Mean Movie Rating (04)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()



```

```python
#Question 2

import re
import numpy as np
from scipy import stats

# 2) Are movies that are newer rated differently than movies that are older? [Hint: Do a median split of year of release to contrast movies in terms of whether they are old or new]

"""
Do:
I extracted the release year of each movie from its title using a regular expression
and added it as a new column in the dataset. Then, using the median split
(median year = 1999), I categorized movies into two groups: new > 1999 and old <1999
I compared thier averages ratings using Lavene's test for hetrogenity of vairances.
Therefore, the Student's t-test was used to compare the two samples since the variances
were hetrogeneous.
Why:
The goal was to examine weather movie age (new vs. old) influences how audiences
rate them. A student t-test was appropriate because it compares the mean ratings between
two independent groups on a continuous scale (0-4). A two-tailed test was chosen
because the question asks whether ratings differ in any direction.
Find:
Levene's test showed no significant difference in variances p = 0.1981 >  = 0.005,
so equal vairances were assumed. The independent samples t-test showed t = 1.605 with df= 398.00
and p(two-tailed) = 0.1092, which is greater than the project significence level  = 0.005.
Answer:
at  = 0.005, the difference in rating between newer and older movies was not statistically
significant. Therefore, our H is kept, indicating that in this dataset, movies
release year does not significantly influence average ratings. In other words,
audiances rated older and newer movies similarly overall.

"""

#Extract release year from movie title
def extract_year(title):
    match = re.search(r'\((\d{4})\)', title)
    return int(match.group(1)) if match else np.nan

movie_cols = movie_stats.index

# Compute year for each film
movie_stats["Year"] = [extract_year(m) for m in movie_stats.index]

# Median split of release year
median_year = movie_stats["Year"].median()
movie_stats["AgeGroup"] = np.where(movie_stats["Year"] >= median_year, "New", "Old")
print("Median release year:", median_year)
print(movie_stats["AgeGroup"].value_counts())

#create two samples
new_means = movie_stats.loc[movie_stats["AgeGroup"]=="New", "Mean Rating"]
old_means = movie_stats.loc[movie_stats["AgeGroup"]=="Old", "Mean Rating"]

s1, s2 = new_means.size, old_means.size
m1, m2 = new_means.mean(), old_means.mean()
sd1, sd2 = new_means.std(ddof=1), old_means.std(ddof=1)
print(f"\nNew n={s1}, mean={m1:.3f}, sd={sd1:.3f}")
print(f"Old n={s2}, mean={m2:.3f}, sd={sd2:.3f}")

# Levenes test for equality of variances
W, p_levene = stats.levene(new_means, old_means, center='median')
if p_levene >= alpha_var:
    print(f"p = {p_levene:.4f}  {alpha_var}  Homogeneity of variances assumed (use Students t-test)")
    equal_var_flag = True
else:
    print(f"p = {p_levene:.4f} < {alpha_var}  Heterogeneity of variances (use Welchs t-test)")
    equal_var_flag = False

# H0: new_means = old_means
# H1: new_means  old_means
# Degrees of freedom
df_t = s1 + s2 - 2
print (f"Students df = {df_t:.2f}")
#Students t-test
t_stat, p_val_two = stats.ttest_ind(new_means, old_means,equal_var=equal_var_flag,nan_policy='omit')

decision = "Significant(drop H0)" if p_val_two < alpha else "Not significant (keep assumption that H0 is true)"
print(f"t = {t_stat:.3f}, p(two-tailed) = {p_val_two:.4g}, alpha = {alpha}")
print("Decision:", decision)


import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))
plt.boxplot(
    [new_means, old_means],
    labels=["New", "Old"],
    patch_artist=True,
    boxprops=dict(facecolor="#66b3ff", alpha=0.7),
    medianprops=dict(color="black", linewidth=1.5),
    whiskerprops=dict(color="gray"),
    capprops=dict(color="gray")
)

plt.title("Comparison of Average Ratings\nNew vs. Old Movies")
plt.ylabel("Average Movie Rating")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Question 10

"""
Do: We applied the KruskalWallis H test to compare viewer ratings across all movies within each franchise,using a significance level of  = 0.005.
    For each franchise, all movies containing the franchise name were grouped together and their rating distributions were compared
    to test whether the overall quality was consistent across movies in the same series.

Why:Each franchise contains multiple independent movies rated on a 04 ordinal scale, Because the data involve more than two samples
    the non-parametric KruskalWallis H test was appropriate.


Find:7 out of 8 franchises (87.5%) showed significant differences in rating distributions (p < 0.005),
     indicating inconsistent quality among their movies.

Answer:Since 7 of the 8 franchises had significant results, we drop our null hypothesis (viewer-rated quality varies significantly
       within most movie franchises).

"""
# create a list with all franchises
franchises = [
    "Star Wars", "Harry Potter", "The Matrix",
    "Indiana Jones", "Jurassic Park",
    "Pirates of the Caribbean", "Toy Story", "Batman"
]

inconsistent=0
results = []

for series in franchises:
  #all films for each serie
   series_cols = [c for c in movie_cols if series.lower() in str(c).lower()]
   #create the samples
   samples = [df_clean[c].dropna() for c in series_cols if df_clean[c].dropna().size >= 10]

   if len(samples) > 1:
      #H0: All movies in the franchise have equal rating distributions.
      # using KruskalWallis H test
        H, p = stats.kruskal(*samples)
      #Calculate the number of inconsistent samples
        is_inconsistent = (p < alpha)
        if is_inconsistent:
            inconsistent += 1
            KW_df=len(samples)-1
        results.append((series, len(samples), H, p, KW_df,"Inconsistent" if is_inconsistent else "Consistent"))

   else:
        results.append((series, len(samples), None, None, "Insufficient movies"))

# For printing purposes
for r in results:
    series, k, H, p,KW_df, status = r
    if H is not None:
        print(f"{series} (k={k} films): H={H:.3f}, p={p:.4g}  {status}")
        print(f"Degrees of freedom (df) = {KW_df}")

    else:
        print(f"{series} (k={k} films): {status}")

print(f"\nTotal inconsistent franchises: {inconsistent} out of {len(franchises)}")

```

```python
from scipy import stats
# Q: Is there a gemder difference in emotional response to movies, as reflicted by agreement with the statment "I have cried during a movie"
# H: there is no difference between male and female participants in how strongly
# they agree with the statement " I Have cried during a movie"
# H: there is a difference between male and female participants in how strongly they agree
# with the statment.
"""
Do:
I examined whether male and female participants differed in their emotional response to movies
using the statment " I have cried during a movie" (column 464). During data inspection, I found an inconistency between the dataset
documentation and the actual responses: several Movie Experiences items (including this one)
had values rating from 1-6 instead of 1-5. Because these values were frquent and
consistent across mubliple variables, they were not treated as a data entery errors.
Therefore, I analyzed the data using the observed range (1-6) to preserve the integrity of
participants' responses. Only valid responses (non missing) were included.
Since the Data are ordinal and the two samples are independet, a two tailed non parametric
Mann-Whitney U test was used to compare the distrbutions.
Why:
The goal was to explore weather there is a gender-based difference in emotional
expressions while whatching movies. The Mann-Whitney U test was appropriate because it does not
assume normality or equal variances and is ideal for comparing two independent samples on an ordinal scale.
A two taild test was chosen because the hypothesis did not precict which gender
would report higher agreement with the statment.
Find:
Female participants (n=806) reported a mean agreement score of 4.215 while male
participants (n=260) reported a mean score of 2.977. The Mann-Whitney U test shwoed
U = 160442 and p = 2.7778183274491205e-40.
Answer:
at  = 0.005, the result was significant, so the null hypothesis is dropped. This
indecates a significant gender difference in emotional responses to movies. Female
participants reported higher agreement whith the statment " I have cried during a movie"
suggesting that females tend to express stronger emotioal reactions while watching movies.
"""

alpha = 0.005
gender_col = df.columns[474]
emotion_col = df.columns[464]

sub = df[df[gender_col].isin([1, 2])]

female_responses = sub.loc[sub[gender_col] == 1, emotion_col].dropna()
male_responses   = sub.loc[sub[gender_col] == 2, emotion_col].dropna()

m1, s1 = female_responses.mean(), female_responses.std(ddof=1)
m2, s2 = male_responses.mean(), male_responses.std(ddof=1)

print("Variable:", emotion_col)
print(f"Female n={female_responses.size}, mean={m1:.3f}, sd={s1:.3f}")
print(f"Male   n={male_responses.size}, mean={m2:.3f}, sd={s2:.3f}")

# MannWhitney U test (two-sided, since were testing for any difference)
U, p = stats.mannwhitneyu(female_responses, male_responses, alternative='two-sided')

print(f"MannWhitney U = {U:.3f}, p = {p:}")
print(f"alpha = {alpha}")

if p < alpha:
    print("Significant  Reject H")
    print("Conclusion: There is a significant gender difference in emotional response (crying during movies).")
else:
    print("Not significant  Fail to reject H")
    print("Conclusion: No significant gender difference in emotional response (crying during movies).")
```
