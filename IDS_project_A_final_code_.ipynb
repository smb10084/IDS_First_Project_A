```python

popularity = df[movie_cols].notna().sum()

mean_rating = df[movie_cols].mean(skipna=True)

movie_stats = pd.DataFrame({"Popularity": popularity, "Mean Rating": mean_rating})

print(movie_stats)
```

```python
import numpy as np
from scipy import stats
alpha =  0.005
alpha_var = 0.005

# 1) Are movies that are more popular (operationalized as having more ratings) rated higher than movies that are less popular? [Hint: You can do
# a median-split of popularity to determine high vs. low popularity movies]
median_popularity = movie_stats["Popularity"].median()
movie_stats["PopularityGroup"] = np.where(movie_stats["Popularity"] >= median_popularity, "High", "Low")
movie_stats["PopularityGroup"].value_counts()

#create two samples
high_means = movie_stats.loc[movie_stats["PopularityGroup"]=="High", "Mean Rating"]
low_means  = movie_stats.loc[movie_stats["PopularityGroup"]=="Low",  "Mean Rating"]

s1, s2 = high_means.size, low_means.size
m1, m2 = high_means.mean(), low_means.mean()
sd1, sd2 = high_means.std(ddof=1), low_means.std(ddof=1)

print("High n =", s1, " | mean =", m1, " | sd =", sd1)
print("Low  n =", s2,  " | mean =", m2,  " | sd =", sd2)


# Levenes test for equality of variances
W, p_levene = stats.levene(high_means, low_means, center='median')
if p_levene >= alpha_var:
    print(f"p = {p_levene:.4f}  {alpha_var}  Homogeneity of variances assumed (use Students t-test)")
    equal_var_flag = True
else:
    print(f"p = {p_levene:.4f} < {alpha_var}  Heterogeneity of variances (use Welchs t-test)")
    equal_var_flag = False

#H : high_means  low_means
#H : high_means > low_means
# Welchs t-test
""" * does not assume homogeneity of variance
    * automatically calculates different degrees of freedom (df) for the two groups
    """
# Degrees of freedom
df_welch = ((sd1**2/s1 + sd2**2/s2)**2) / (((sd1**2/s1)**2)/(s1-1) + ((sd2**2/s2)**2)/(s2-1))
print(f"Welchs df = {df_welch:.2f}")
# If there are missing values (NaN), ignore them and continue the calculation without stopping "nan_policy='omit'"
t_stat, p_val = stats.ttest_ind(high_means, low_means, equal_var=equal_var_flag, nan_policy='omit')
# convert P-value to one-tailed (High > Low)
p_val_one = p_val/ 2 if t_stat > 0 else 1.0
decision = "Significant(drop H0)" if p_val_one < alpha else "Not significant (keep assumption that H0 is true)"
print(f"t = {t_stat:.3f}, p(one-tailed) = {p_val_one:.4g}, alpha = {alpha}")
print("Decision:", decision)



import matplotlib.pyplot as plt

data = [high_means, low_means]
labels = ['High Popularity', 'Low Popularity']

plt.figure(figsize=(7,5))
plt.boxplot(data, tick_labels=labels, patch_artist=True)
plt.title("Comparison of Mean Ratings by Popularity Group", fontsize=14)
plt.ylabel("Mean Movie Rating (04)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()



```

```python
#Question 2

import re
import numpy as np
from scipy import stats

# 2) Are movies that are newer rated differently than movies that are older? [Hint: Do a median split of year of release to contrast movies in terms of whether they are old or new]


#Extract release year from movie title
def extract_year(title):
    match = re.search(r'\((\d{4})\)', title)
    return int(match.group(1)) if match else np.nan

movie_cols = movie_stats.index

# Compute year for each film
movie_stats["Year"] = [extract_year(m) for m in movie_stats.index]

# Median split of release year
median_year = movie_stats["Year"].median()
movie_stats["AgeGroup"] = np.where(movie_stats["Year"] >= median_year, "New", "Old")
print("Median release year:", median_year)
print(movie_stats["AgeGroup"].value_counts())

#create two samples
new_means = movie_stats.loc[movie_stats["AgeGroup"]=="New", "Mean Rating"]
old_means = movie_stats.loc[movie_stats["AgeGroup"]=="Old", "Mean Rating"]

s1, s2 = new_means.size, old_means.size
m1, m2 = new_means.mean(), old_means.mean()
sd1, sd2 = new_means.std(ddof=1), old_means.std(ddof=1)
print(f"\nNew n={s1}, mean={m1:.3f}, sd={sd1:.3f}")
print(f"Old n={s2}, mean={m2:.3f}, sd={sd2:.3f}")

# Levenes test for equality of variances
W, p_levene = stats.levene(new_means, old_means, center='median')
if p_levene >= alpha_var:
    print(f"p = {p_levene:.4f}  {alpha_var}  Homogeneity of variances assumed (use Students t-test)")
    equal_var_flag = True
else:
    print(f"p = {p_levene:.4f} < {alpha_var}  Heterogeneity of variances (use Welchs t-test)")
    equal_var_flag = False

# H0: new_means = old_means
# H1: new_means  old_means
# Degrees of freedom
df_t = s1 + s2 - 2
print (f"Students df = {df_t:.2f}")
#Students t-test
t_stat, p_val_two = stats.ttest_ind(new_means, old_means,equal_var=equal_var_flag,nan_policy='omit')

decision = "Significant(drop H0)" if p_val_two < alpha else "Not significant (keep assumption that H0 is true)"
print(f"t = {t_stat:.3f}, p(two-tailed) = {p_val_two:.4g}, alpha = {alpha}")
print("Decision:", decision)


import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))
plt.boxplot(
    [new_means, old_means],
    labels=["New", "Old"],
    patch_artist=True,
    boxprops=dict(facecolor="#66b3ff", alpha=0.7),
    medianprops=dict(color="black", linewidth=1.5),
    whiskerprops=dict(color="gray"),
    capprops=dict(color="gray")
)

plt.title("Comparison of Average Ratings\nNew vs. Old Movies")
plt.ylabel("Average Movie Rating")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#Question 4

#4) What proportion of movies are rated differently by male and female viewers?
gender_col = df.columns[474]
movie_cols = df.columns[:400]


#Define two counters
total_tested = 0
significant_count = 0

for movie in movie_cols:
    #create two samples for each film
    female_ratings = df.loc[df[gender_col] == 1, movie].dropna()
    male_ratings   = df.loc[df[gender_col] == 2, movie].dropna()

    if len(female_ratings) > 10 and len(male_ratings) > 10:
        total_tested += 1
        # H0: female_ratings  = male_ratings for each film
        # H1: female_ratings  male_ratings  for each film
        #Mann-Whitney U test for each film
        u_stat, p_val = stats.mannwhitneyu(female_ratings, male_ratings, alternative='two-sided')
        if p_val < alpha:
            significant_count += 1


proportion = significant_count / total_tested
print(f"Movies tested: {total_tested}")
print(f"Significant differences: {significant_count}")
print(f"Proportion = {proportion:.2%}")

import matplotlib.pyplot as plt

significant = significant_count
not_significant = total_tested - significant

labels = ['Significant difference', 'No significant difference']
sizes = [significant, not_significant]
colors = ['#ff9999','#66b3ff']

plt.figure(figsize=(5,5))
plt.pie(sizes, labels=labels, autopct='%1.2f%%', startangle=140, colors=colors)
plt.title('Proportion of Movies Rated Differently by Male and Female Viewers')
plt.show()
```

```python
#Question 8
movie_cols = df.columns[:400]
social_col = df.columns[476]
alpha = 0.005

df_clean = df.drop(df[df[social_col] == -1.0].index)

df_social = df_clean[df_clean[social_col] == 0]
df_alone = df_clean[df_clean[social_col] == 1]

sig_count = 0
significant_movies = []
movies_tested = 0

# 4. Loop through all 400 movies
for movie in movie_cols:

    social_ratings = df_social[movie].dropna()
    alone_ratings = df_alone[movie].dropna()

    if len(social_ratings) > 0 and len(alone_ratings) > 0:
        movies_tested += 1

        statistic, p_value = stats.mannwhitneyu(
            social_ratings,
            alone_ratings,
            alternative='two-sided'
        )

        if p_value < alpha:
            sig_count += 1
            significant_movies.append(movie)


#Calculate the final proportion
proportion = sig_count / len(movie_cols)

print(f"--- Test Results ---")
print(f"Total movies tested: {movies_tested}")
print(f"Movies with a significant effect (p < {alpha}): {sig_count}")
print(f"Proportion of movies exhibiting a 'social watching' effect: {proportion:.4f} (or {proportion:.2%})\n")
```

```python
# Question 10

# create a list with all franchises
franchises = [
    "Star Wars", "Harry Potter", "The Matrix",
    "Indiana Jones", "Jurassic Park",
    "Pirates of the Caribbean", "Toy Story", "Batman"
]

inconsistent=0
results = []

for series in franchises:
  #all films for each serie
   series_cols = [c for c in movie_cols if series.lower() in str(c).lower()]
   #create the samples
   samples = [df_clean[c].dropna() for c in series_cols if df_clean[c].dropna().size >= 10]

   if len(samples) > 1:
      #H0: All movies in the franchise have equal rating distributions.
      # using KruskalWallis H test
        H, p = stats.kruskal(*samples)
      #Calculate the number of inconsistent samples
        is_inconsistent = (p < alpha)
        if is_inconsistent:
            inconsistent += 1
            KW_df=len(samples)-1
        results.append((series, len(samples), H, p, KW_df,"Inconsistent" if is_inconsistent else "Consistent"))

   else:
        results.append((series, len(samples), None, None, "Insufficient movies"))

# For printing purposes
for r in results:
    series, k, H, p,KW_df, status = r
    if H is not None:
        print(f"{series} (k={k} films): H={H:.3f}, p={p:.4g}  {status}")
        print(f"Degrees of freedom (df) = {KW_df}")

    else:
        print(f"{series} (k={k} films): {status}")

print(f"\nTotal inconsistent franchises: {inconsistent} out of {len(franchises)}")

```

```python
import matplotlib.pyplot as plt
import pandas as pd

df_franchise = pd.DataFrame(results, columns=["Franchise", "NumMovies", "H", "p", "df", "Status"])
df_franchise = df_franchise.dropna(subset=["p"])

plt.figure(figsize=(8,5))
bars = plt.bar(df_franchise["Franchise"], df_franchise["p"],
               color=df_franchise["Status"].map({"Inconsistent": "#ff9999", "Consistent": "#66b3ff"}))

plt.axhline(0.005, color='red', linestyle='--', label=' = 0.005')
plt.yscale('log')
plt.ylabel("p-value (log scale)")
plt.title("Franchise Consistency Based on KruskalWallis Test")
plt.xticks(rotation=45, ha='right')
plt.legend()

for bar, p in zip(bars, df_franchise["p"]):
    plt.text(bar.get_x() + bar.get_width()/2, p, f"{p:.3g}", ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
from scipy import stats
# Q: Is there a gemder difference in emotional response to movies, as reflicted by agreement with the statment "I have cried during a movie"
# H: there is no difference between male and female participants in how strongly
# they agree with the statement " I Have cried during a movie"
# H: there is a difference between male and female participants in how strongly they agree
# with the statment.

alpha = 0.005
gender_col = df.columns[474]
emotion_col = df.columns[464]

sub = df[df[gender_col].isin([1, 2])]

female_responses = sub.loc[sub[gender_col] == 1, emotion_col].dropna()
male_responses   = sub.loc[sub[gender_col] == 2, emotion_col].dropna()

m1, s1 = female_responses.mean(), female_responses.std(ddof=1)
m2, s2 = male_responses.mean(), male_responses.std(ddof=1)

print("Variable:", emotion_col)
print(f"Female n={female_responses.size}, mean={m1:.3f}, sd={s1:.3f}")
print(f"Male   n={male_responses.size}, mean={m2:.3f}, sd={s2:.3f}")

# MannWhitney U test (two-sided, since were testing for any difference)
U, p = stats.mannwhitneyu(female_responses, male_responses, alternative='two-sided')

print(f"MannWhitney U = {U:.3f}, p = {p:}")
print(f"alpha = {alpha}")

if p < alpha:
    print("Significant  Reject H")
    print("Conclusion: There is a significant gender difference in emotional response (crying during movies).")
else:
    print("Not significant  Fail to reject H")
    print("Conclusion: No significant gender difference in emotional response (crying during movies).")
```
